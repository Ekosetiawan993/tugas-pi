### Function for evaluation

```python
from sklearn.metrics import mean_squared_error
```

```python
def model_evaluation(model, model_name):
    train_pred = model.predict(train_x_scl)
    test_pred = model.predict(test_x_scl)
    
    train_mse = mean_squared_error(train_y, train_pred)
    test_mse = mean_squared_error(test_y, test_pred)
    
    train_rmse = mean_squared_error(train_y, train_pred, squared=False)
    test_rmse = mean_squared_error(test_y, test_pred, squared=False)
    
    print(f'Model: {model_name}')
    print(f'Train MSE    : {train_mse}')
    print(f'Train RMSE   : {train_rmse}')
    print(f'Test MSE     : {test_mse}')
    print(f'Test RMSE    : {test_rmse}')
```

```python
def evaluation_2(model, model_name):
    train_pred = model.predict(train_x_scl)
    test_pred = model.predict(test_x_scl)
    
    train_mse = mean_squared_error(train_y, train_pred)
    test_mse = mean_squared_error(test_y, test_pred)
    
    train_rmse = mean_squared_error(train_y, train_pred, squared=False)
    test_rmse = mean_squared_error(test_y, test_pred, squared=False)
    
    print(f'Model: {model_name}')
    print(f'Train MSE    : {train_mse}')
    print(f'Train RMSE   : {train_rmse}')
    print(f'Test MSE     : {test_mse}')
    print(f'Test RMSE    : {test_rmse}')
```

### Ridge Linear Regression

```python
from sklearn.linear_model import Ridge
from sklearn.model_selection import GridSearchCV
```

```python
ridge = Ridge()

parameters = {'alpha' : [1e-15, 1e-10, 1e-8, 1e-4, 1e-3, 1e-2, 1, 5, 10, 20]}
```

```python
ridge_regressor = GridSearchCV(ridge, parameters, scoring='neg_root_mean_squared_error', cv=5)
ridge_regressor.fit(train_x_scl, train_y)
```

```python
print(ridge_regressor.best_params_)
print(-ridge_regressor.best_score_)
```

```python
ridge_reg = Ridge(alpha=0.01)
ridge_reg.fit(train_x_scl, train_y)
```

```python
evaluation(ridge_reg, 'Ridge Linear Regression')
```

### Multi Layer Regressor

```python
from sklearn.neural_network import MLPRegressor
```

```python
mlp_reg = MLPRegressor(hidden_layer_sizes=(150,), random_state=1, max_iter=500)
```

```python
mlp_reg.fit(train_x_scl, train_y.values.ravel())
```

```python
evaluation(mlp_reg, 'Multi Later Perceptron 500')
```

```python
evaluation(mlp_reg, 'Multi Later Perceptron 500, hl = 150')
```

```python
from sklearn.neural_network import MLPRegressor
```

```python
hidden_layer = []
mlp_rmse_collection = []

for x in range(100, 260, 10):
    mlp = MLPRegressor(hidden_layer_sizes=(x,), random_state=1, max_iter=500)
    mlp.fit(train_x_scl, train_y.values.ravel())
    hidden_layer.append(x)
    prediction = mlp.predict(test_x_scl)
    mlp_rmse = mean_squared_error(test_y, prediction, squared=False)
    mlp_rmse_collection.append(mlp_rmse)
```

```python
plt.figure(figsize=(10, 8))
sns.lineplot(x=hidden_layer, y=mlp_rmse_collection)
plt.title('RMSE MLP Regressor vs n_estimators')
plt.xlabel('Number of hidden layers')
plt.ylabel('Test RMSE')
plt.savefig('mlp_rmse.png')
plt.show()
```

```python
final_mlp_reg = MLPRegressor(hidden_layer_sizes=(110,), random_state=1, max_iter=500)
```

```python
final_mlp_reg.fit(train_x_scl, train_y.values.ravel())
```

```python
model_evaluation(final_mlp_reg, 'Multi Later Perceptron Regressor(hidden_layer=110)')
```
