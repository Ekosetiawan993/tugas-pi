### Split

```python
from sklearn.model_selection import train_test_split
```

```python
x = data_5.drop('Age', axis=1)
```

```python
y = data_5.Age
```

```python
train_x, test_x, train_y, test_y = train_test_split(x, y, test_size=0.25, random_state=42)
```

```python
train_x.head()
```

### Normalize data

```python
from sklearn.preprocessing import StandardScaler
```

```python
scl = StandardScaler()
```

```python
train_x = scl.fit_transform(train_x)
```

```python
test_x = scl.transform(test_x)
```

```python
train_x
```

```python
test_x
```

### Make Models

#### Linear Regression

```python
from sklearn.linear_model import LinearRegression
```

```python
lin_reg = LinearRegression()
```

```python
lin_reg.fit(train_x, train_y)
```

#### evaluation

```python
from sklearn.metrics import mean_squared_error
```

```python
def evaluation(model, model_name):
    train_pred = model.predict(train_x)
    test_pred = model.predict(test_x)
    train_rmse = mean_squared_error(train_y, train_pred, squared=False)
    test_rmse = mean_squared_error(test_y, test_pred, squared=False)
    
    print(f'Model: {model_name}')
    print(f'Train RMSE    : {train_rmse}')
    print(f'Test RMSE     : {test_rmse}')
```

```python
evaluation(lin_reg, 'Linear Regression')
```

### Ridge Linear Regression

```python
from sklearn.linear_model import Ridge
from sklearn.model_selection import GridSearchCV
```

```python
ridge = Ridge()

parameters = {'alpha' : [1e-15, 1e-10, 1e-8, 1e-4, 1e-2, 1, 5, 10, 20]}

```

```python
ridge_regressor = GridSearchCV(ridge, parameters, scoring='neg_root_mean_squared_error', cv=5)
ridge_regressor.fit(train_x, train_y)
```

```python
print(ridge_regressor.best_params_)
print(-ridge_regressor.best_score_)
```

```python
ridge_reg = Ridge(alpha=10)
ridge_reg.fit(train_x, train_y)
```

```python
test_pred = ridge_reg.predict(test_x)
```

```python
ridge_rmse = mean_squared_error(test_y, test_pred, squared=False)
```

```python
ridge_rmse
```
